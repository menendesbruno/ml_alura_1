```python
import pandas as pd
import graphviz
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split, cross_validate, KFold, StratifiedKFold, GroupKFold
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.dummy import DummyClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.pipeline import Pipeline
```

```python
uri = "https://gist.githubusercontent.com/guilhermesilveira/e99a526b2e7ccc6c3b70f53db43a87d2/raw/1605fc74aa778066bf2e6695e24d53cf65f2f447/machine-learning-carros-simulacao.csv"

dados = pd.read_csv(uri).drop(columns=["Unnamed: 0"], axis=1)

dados.head()
```

```python
renome = {
    'km_por_ano': 'km_ano',
    'idade_do_modelo': 'idade',
    'price': 'preco',
    'sold': 'vendido'
}
dados = dados.rename(columns=renome)
dados.head()
```

```python

dados.head()
```

```python
soma_modelo = np.random.randint(-2, 3, size=10000)
dados['modelo'] = dados.idade + soma_modelo
```

```python
dados.head()
dados.modelo.unique()
dados['modelo'] = dados.modelo + abs(dados.modelo.min()) + 1
dados.modelo.unique()
```

```python
x = dados[['preco', 'idade', 'km_ano']]
y = dados['vendido']
```

```python
SEED = 158020
np.random.seed = 158020

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, test_size = 0.25, stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

modelo = LinearSVC(random_state = SEED)
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

```python
dummy = DummyClassifier()

dummy.fit(treino_x, treino_y)
previsoes = dummy.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

```python
dummy = DummyClassifier()

dummy.fit(treino_x, treino_y)

acuracia = dummy.score(teste_x, teste_y) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

```python
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25, stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(raw_treino_x), len(raw_teste_x)))

scaler = StandardScaler()
scaler.fit(raw_treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)

modelo = SVC(gamma='auto')
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

```python
# cross validation
modelo = DecisionTreeClassifier(max_depth=2)
results = cross_validate(modelo, x, y, cv = 5, return_train_score=False)
print(results['test_score'])
media = results['test_score'].mean()
desvio = results['test_score'].std()
print(f"Acuracy com cross validation 5: {round((media - 2 * desvio)*100, 2)} e {round((media + 2 * desvio)*100, 2)} com media de {round(media * 100, 2)}")
```

# Aleatoridade com Cross validate

```python
SEED = 301

cv = KFold(n_splits=10, shuffle=True)
modelo = DecisionTreeClassifier(max_depth=2)
results = cross_validate(modelo, x, y, cv = cv, return_train_score=False)
print(results['test_score'])
media = results['test_score'].mean()
desvio = results['test_score'].std()
print(f"Acuracy com cross validation 5: {round((media - 2 * desvio)*100, 2)} e {round((media + 2 * desvio)*100, 2)} com media de {round(media * 100, 2)}")
```

```python
def imprime_results(results, cv):
    print(results['test_score'])
    media = results['test_score'].mean()
    desvio = results['test_score'].std()
    print(f"Acuracy com cross validation {cv}: {round((media - 2 * desvio)*100, 2)} e {round((media + 2 * desvio)*100, 2)} com media de {round(media * 100, 2)}")
```

```python
cv = GroupKFold(n_splits=10)
modelo = SVC()
results = cross_validate(modelo, x_azar, y_azar, groups=dados.modelo,cv = cv, return_train_score=False)
imprime_results(results, cv)
```

```python
scaler = StandardScaler()
scaler.fit(x_azar)
x_azar_escalado = scaler.transform(x_azar)
results = cross_validate(modelo, x_azar_escalado, y_azar, groups=dados.modelo,cv = cv, return_train_score=False)
imprime_results(results, cv)
```

```python
scaler = StandardScaler()
modelo = SVC()
pipeline = Pipeline([('transformacao', scaler), ('estimador', modelo)])
cv = GroupKFold(n_splits=10)
results = cross_validate(pipeline, x_azar, y_azar, groups=dados.modelo,cv = cv, return_train_score=False)
imprime_results(results, cv)
```
